#DIAGNÓSTICO DOS DADOS

## Diagnóstico do Problema

Paradas inesperadas em equipamentos industriais causam perdas operacionais e
custos elevados de manutenção. Este projeto propõe o uso de manutenção
preditiva baseada em dados de sensores para prever falhas antes que ocorram,
garantindo operação contínua (Zero-Downtime).

Sensores monitorados:
- Temperatura
- Rotação
- Torque
- Desgaste da ferramenta
